# EXPLORE DATA

```python
numeric_features=[feature for feature in df2.columns if df2[feature].dtype !='O']
categorical_features=[feature for feature in df2.columns if df2[feature].dtype =='O']

print("We have {} Numerical Columns:{}".format(len(numeric_features),numeric_features))
print("We have{} Categorical Columns:{}".format(len(categorical_features),categorical_features))
```

```python
for columns in categorical_features:
    print(df2[columns].value_counts(normalize=True)*100)
    print()
    print("________________________________________________")
    print()
```

# Data Visualisation

```python
plt.figure(figsize=(15,16))
plt.suptitle("Univariate Analysis of Numeric Features ", fontweight="bold", fontsize=20 , alpha =0.8 , y= 1.)

for i in range(len(numeric_features)):
    plt.subplot(3, 3, i + 1)
    sns.kdeplot(df2[numeric_features[i]],shade=True, color="r")
    plt.xlabel(numeric_features[i])
    plt.tight_layout()   
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.suptitle("Univariate Analysis of categorical feature ", fontweight="bold", fontsize=20, alpha=0.8,y=1.)
category=['category_by_Gender', 'Category']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df2[category[i]],hue=df2[category[i]],palette="pastel",legend=False)
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
