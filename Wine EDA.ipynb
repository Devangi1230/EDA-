# Explore analysis

```python
numeric_col= [col for col in df.columns if df[col].dtype!='0']
print("We have {} Numeric Columns :{}".format (len(numeric_col),numeric_col))
```

```python
for cols in numeric_col:
    print(df[cols].value_counts(normalize=True)*100)
    print('______________________________________________________')
```

# Data Visualisation

```python
plt.figure(figsize=(15,16))
plt.suptitle("Univariate Analysis of Numeric Columns",fontsize=20, fontweight='bold',y=1.)

for i in range(0,len(numeric_col)):
    plt.subplot(4,3,i+1)
    sns.kdeplot(x=df[numeric_col[i]],shade=True,color='r')
    plt.xlabel(numeric_col[i])
    plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,16))
plt.suptitle("Univariate Analysis of Numeric Column outlier",fontsize=20, fontweight='bold',y=1.)

for i in range(0,len(numeric_col)):
    plt.subplot(4,3,i+1)
    sns.boxplot(y=df[numeric_col[i]],color='skyblue')
    plt.xlabel(numeric_col[i])
    plt.tight_layout()
plt.show()
```

```python
df["quality"].value_counts().sort_index().plot(kind="bar",figsize=(10,5))
plt.title("Distribution of Wine Quality Ratings", fontsize=16, fontweight='bold')
plt.xlabel("Quality Score")
plt.ylabel("Number of Wines")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.show()
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm',square=True,linewidths=0.5)
plt.title("Correalation of Wine Features Through heatmap", fontsize=16, fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
features = ['alcohol', 'volatile acidity', 'density']
```

```python
grouped = df.groupby('quality')[features].mean().reset_index()
grouped
```

```python
melted = grouped.melt(id_vars='quality', var_name='Feature', value_name='Average')
melted
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='quality', y='Average', hue='Feature', data=melted)
plt.title("Average Feature Values by Wine Quality")
plt.xlabel("Quality Score")
plt.ylabel("Average Value")
plt.legend(title="Feature")
plt.show()
```
